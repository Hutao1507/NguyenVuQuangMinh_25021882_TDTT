```python
#W5A1
def Max_Of_Two(a, b):
    if int(a) != a or int(b) != b:
        return "Vui lòng nhập số nguyên!"
    else:
        max = a
        if b > a:
            max = b
        return max
    
a, b = input("Nhập hai số nguyên a, b:").split()
print(Max_Of_Two(a, b))
```

```python
#W5A2
def Swap(a, b) -> Tuple(int, int):
    if int(a) != a or int(b) != b:
        return "Vui lòng nhập số nguyên!"
    else:
        return b,a

a, b = input("Nhập hai số nguyên a, b:").split()
print(Swap(a, b))
```

```python
#W5A3
def So_nguyen_to(a):
    if int(a) != a:
        return "Vui lòng nhập số nguyên!"
    if a < 2:
        return "Không phải số nguyên tố"
    else:
        for i in range(2, int(a**0.5) + 1):
            if a % i == 0:
                return "Không phải số nguyên tố"    
        return "Là số nguyên tố"
    
a= int(input("Nhập số nguyên a:"))
print(So_nguyen_to(a))
```

```python
#W5A4
def so_hoan_hao(n):
    s=[]
    sum=0
    if int(n) != n:
        return "Vui lòng nhập số nguyên!"
    if n < 0:
        return "Vui lòng nhập số nguyên dương!"
    for i in range(1, n):
        if n % i == 0:
            s.append(int(i))
    for gia_tri in range(0, len(s)):
        sum += s[gia_tri]
    if sum == n:
        return "Là số hoàn hảo"
    else:
        return "Không phải số hoàn hảo"
    
print(so_hoan_hao(5))
        
```

```python
#W5A5
def Tim_so_giong_(list, k:int):
    for STT, Value in enumerate(list):
        if Value == k:
            return Value
    return -1

nhap = input("Nhập dãy số với số cuối là số nguyên dương k cách nhau bởi dấu cách:").split() 
k = int(nhap[-1])
list = nhap[:-1]
for i in nhap[:-1]:
    list.append(int(i))
print(Tim_so_giong_(list, k))
    
```

```python
#W5A6
def giai_thua(n):
    if n < 0:
        return "False"
    if n == 0 or n == 1:
        return 1
    else:
        gt = 1
        for i in range(2, n + 1):
            gt *= i
        return gt

try:       
    n = int(input("Nhập số nguyên dương n:"))
    if giai_thua(n) == "False2":
        print("Vui lòng nhập số nguyên dương!")
    else:
        print("giai thừa của", n, "là:", giai_thua(n))
except ValueError:
    print("Vui lòng nhập số nguyên dương")
```

```python
#W5A7
def calculator(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "Không thể chia cho 0!"
        else:
            return a / b
    else:
        return "Phép toán không hợp lệ!"

try:
    a,b = map(float, input("Nhập hai số thực a, b:").split())
    operator = input("Nhập phép toán (+, -, *, /):")
    print("Kết quả:", calculator(a, b, operator))
except ValueError:
    print("Vui lòng nhập số hợp lệ!")
```

```python
#W5A8
def hamming_distance(a,b):
    xor = a^b
    cout=0
    for i in bin(xor):
        if i=='1':
            cout+=1
    return cout
try:
    a, b = map(int, input("Nhập hai số nguyên a, b:").split())    
    assert a>=0 and b>=0
    print("Khoảng cách Hamming giữa", a, "và", b, "là:", hamming_distance(a,b))
except ValueError:
    print("Vui lòng nhập số nguyên")
except AssertionError:
    print("Vui lòng nhập số nguyên dương")
```

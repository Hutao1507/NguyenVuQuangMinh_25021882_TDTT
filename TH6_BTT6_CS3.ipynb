```python
#W6A1
n = input("Nhập chuỗi số nguyên:").split()
m = []
for var in n:
    try:
        var = int(var)
        if var not in m:
            m.append(var)
    except ValueError:
        print("Nhập số nguyên")
        break
print(m)   
```

```python
#W6A2
n = input("Nhập dãy số nguyên:").split()
sum_list = []
sum = 0 
for var in n:
    try:
        var = int(var)
        sum += int(var)
        sum_list.append(sum)
    except ValueError:
        print("Dãy chỉ được chứa số nguyên")
        break
    
print(sum_list)
```

```python
#W6A3
n = input("Nhập dãy tuple:").split()
k = int(input("Nhập số k:"))  
start = []
end = []
for i in range(len(n)):
    try:
        var = int(n[i])
        if i > k-1:
            start.append(var)
        else:
            end.append(var)
    except ValueError:
        print("Các giá trị nhập vào phải là số nguyên")
        break
print(tuple(start + end))        
        
```

```python
#W6A4
input_str = input("Nhập các chuỗi key-value đầu vào").split()
list = {}
for pair in input_str:
    if ":" not in pair:
        print(f"Cặp {pair} sai cú pháp")
        continue
    key,value = pair.split(":", 1)
    key.strip()
    value.strip()
    if not key: # Nếu key rỗng thì not là True
        continue
    if not  value:
        continue
    if key not in list:
        list[key] = [value]
    else:
        list[key].append(value)
        
print(list)
```

```python
#W6A5
list = input("Nhập một dãy só nguyên").split()
dict = {'positives':[], 'negatives':[], 'zeros':[]}
for number in list:
    try:
        number = int(number)
        if number > 0:
            dict['positives'].append(number)
        elif number < 0:
            dict['negatives'].append(number)
        else:
            dict['zeros'].append(number)       
    except ValueError:
        print(f"Số {number} không phải là số nguyên")
        continue
print(dict)
```

```python
#W6A8
n = input("Nhập dãy các xâu:")
n = n.strip('[]')
dict = {}
if not n:
    pass
else:
    n = n.split(",")
    for string in n:
        string = string.strip().strip('""').strip()
        if string not in dict:
            dict[string] = 1
        else:
            dict[string] += 1
print(dict)
```

```python
#W6A9
dict1 = eval(input("Nhập dict thứ nhất: "))
dict2 = eval(input("Nhập dict thứ hai: "))
for key, value in dict2.items():
    if key in dict1:
        dict1[key] += value
    else:
        dict1[key] = value

result = dict(sorted(dict1.items()))
print(result)
```

```python
#W6A10
n = eval(input('Nhập dãy số nguyên:'))
k = int(input("Nhập số nguyên dương k:"))
result = []
for index1 in range(len(n)):
    for index2 in range(index1+1, len(n)):
        if n[index1] + n[index2] == k:
            result.append((n[index1], n[index2]))
        else:
            pass
print(sorted(result))
```

```python
#W6A11
n = eval(input("Nhập vào tuple các số nguyên:"))
odd = []
even = []
try:
    for value in n:
        if int(value) != value:
            raise ValueError
except ValueError:
    print(f"Số {value} không phải là số nguyên")
else:
    for number in n:
        if number%2 == 0:
            even.append(number)
        else:
            odd.append(number)
    print(f"Tuple số chẵn là: {tuple(even)} \n Tuple số lẻ là {tuple(odd)}")
    
            
```

```python
#W6A12
n = eval(input("Nhập một list các số nguyên:"))
my_dict = {}

for number in n:
    if number in my_dict:
        my_dict[number] += 1
    else:
        my_dict[number] = 1

sort_dict = dict(sorted(my_dict.items()))
max_key = 0
max_value = 0

for key, value in sort_dict.items():
    if value > max_value:
        max_key = key
        max_value = value
    else:
        pass
    
print(f"Giá trị xuất hiện nhiều nhất là {max_key}, số lần xuất hiện {max_value}")
    
```

```python
def luy_thua(x, n):
    if n == 0:
        return 1
    return x * luy_thua(x, n - 1)

print(luy_thua(2, 3))
```

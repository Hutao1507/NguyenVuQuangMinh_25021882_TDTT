```python
#W7A1
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = input().split()
x = input()
print(binary_search(arr, x))
```

```python
#W7A3
def buble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - 1 - i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min = i
        for j in range(i + 1, n):
            if arr[j] < arr[min]:
                min = j
        arr[i], arr[min] = arr[min], arr[i]
    return arr
            

arr = input().split()
print(selection_sort(arr))
```

```python
#W7A4
def count_appear(arr):
    dict = {}
    for value in arr:
        if value not in dict:
            dict[value] = 1
        else:
            dict[value] += 1
    max_key = 0
    max_time = 0
    for key, value in dict.items():
        if value > max_time:
            max_key = key
            max_time = value
    return max_key, max_time

arr = input().split()
print(f'{count_appear(arr)[0]} xuat hien nhieu nhat, som nhat, {count_appear(arr)[1]} lan')
```

```python
#W7A7
def find_length(arr):
    if not arr:
        return []
    end = []
    for value in arr:
        low = 0
        high = len(end)
        while low < high:
            mid = (high + low)//2
            if end[mid] < value:
                low = mid + 1
            else: 
                high = mid
        if low == len(end):
            end.append(value)
        else:
            end[low] = value
    return len(end)
arr = eval(input())
print(find_length(arr))
```

```python
#W7A8
def find_length(interval, query):
    idx_query = []
    min_length = [float('inf')] * len(query)
    for i in range(len(query)):
        idx_query.append([query[i], i])
    srt_idx_query = sorted(idx_query)
    sort_query = sorted(query)
    def find_idx_start(s):
        low = 0
        high = len(sort_query) - 1
        ans = len(sort_query)
        while low <= high:
            mid = (low + high)//2
            if sort_query[mid] >= s:
                ans = mid
                high = mid - 1
            else:
                low = mid + 1
        return ans
    def find_idx_end(e):
        low = 0
        high = len(sort_query) - 1
        ans = len(sort_query)
        while low <= high:
            mid = (low + high)//2
            if sort_query[mid] <= e:
                ans = mid
                low = mid + 1
            else:
                high = mid - 1
        return ans
    for l, r in interval:
        start = find_idx_start(l)
        end = find_idx_end(r)
        length = r - l + 1
        for i in range(start, end + 1):
            if min_length[i] > length:
                min_length[i] = length
    final_results = [-1] * len(query)
    for i in range(len(query)):
        original_idx = srt_idx_query[i][1]
        if min_length[i] != float('inf'):
            final_results[original_idx] = min_length[i]
    return final_results

interval = eval(input())
query = eval(input())
print(find_length(interval, query))
        
```

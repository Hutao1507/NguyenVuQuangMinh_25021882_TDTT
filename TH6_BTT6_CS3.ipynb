```python
#W6A1
n = input("Nhập chuỗi số nguyên:").split()
m = []
for var in n:
    try:
        var = int(var)
        if var not in m:
            m.append(var)
    except ValueError:
        print("Nhập số nguyên")
        break
print(m)   
```

```python
#W6A2
n = input("Nhập dãy số nguyên:").split()
sum_list = []
sum = 0 
for var in n:
    try:
        var = int(var)
        sum += int(var)
        sum_list.append(sum)
    except ValueError:
        print("Dãy chỉ được chứa số nguyên")
        break
    
print(sum_list)
```

```python
#W6A3
n = input("Nhập dãy tuple:").split()
k = int(input("Nhập số k:"))  
p = 0
list = []
for i in range(len(n)):
    try:
        var = int(n[i])
        if i < k:
            list.append(var)
            p = p+1
        else:
            list.insert(len(list)-p-1, var)
    except ValueError:
        print("Các giá trị nhập vào phải là số nguyên")
        break
print(p)
print(list)
        
        
```
